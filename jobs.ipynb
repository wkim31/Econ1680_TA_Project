# Save to json file

```python
import requests
import json

API_ID = 'c351eda7'
API_KEY = '39c0ab4ac8fc94d4f0dcf3f1d95411d7'
base_endpoint = 'http://api.adzuna.com/v1/api/jobs/us/search/'
num_pages = 30

with open('job_data.json', 'w') as file:
    for page in range(1, num_pages + 1):
        endpoint = f'{base_endpoint}{page}'
        response = requests.get(f'{endpoint}?app_id={API_ID}&app_key={API_KEY}&content-type=application/json')
        
        if response.status_code == 200:
            job_data = response.json()
            file.write(json.dumps(job_data))  # Write the job data to the file
            print(f"Page {page} - Job Data saved.")
        else:
            print(f"Error: {response.status_code}")
            print(response.text)  #if error, print response
```
